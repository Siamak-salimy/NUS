```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/drive/MyDrive/NUS/diabetes.csv')
```

```python

X = data.drop('Outcome', axis=1)
y = data['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
new_data = pd.read_csv('/content/drive/MyDrive/NUS/newdata.csv')
```

```python

X_new = new_data.drop('Outcome', axis=1)
y_pred_new = clf.predict(X_new)
new_data['predicted_label'] = y_pred_new
print ("predicted_label For input data is : ",y_pred_new)
```
